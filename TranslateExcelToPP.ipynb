```python
year = 2021
month = 3
day = 1
types = {
    0: 'ОбяЕ',
    2: 'ЭкзЕ',
    4: 'Кчаю',
    6: 'Вкус',
    8: 'Проч',
    10: 'Одеж',
    12: 'Кафе',
    14: 'Техн',
    16: 'Игры',
}
monthDic = {
    'Январь': 1,
    'Февраль': 2,
    'Март': 3,
    'Апрель': 4,
    'Май': 5,
    'Июнь': 6,
    'Июль': 7,
    'Август': 8,
    'Сентябрь': 9,
    'Октябрь': 10,
    'Ноябрь': 11,
    'Декабрь': 12,
}

def str2arr(sum, string):
  if not sum:
    return None, None, None, None, None
  arrstr = string.split(',')
  arrsum = str(sum)
  if arrsum[0] == '=':
    arrsum = arrsum[1:]
  arrsum = arrsum.split('+')
  numstr = []
  isWeight = []

  iswrong = len(arrstr) != len(arrsum)
  if iswrong:
    return string, sum, [0,], iswrong
  for i in range(len(arrsum)):
    isWeight.append(False)
    bisum = False
    temp = arrsum[i].split('*')
    if len(temp) == 2:
      bisum = True
      if float(temp[0]) > float(temp[1]):
        arrsum[i] = float(temp[0])
        numstr.append(float(temp[1]))
      else:
        arrsum[i] = float(temp[1])
        numstr.append(float(temp[0]))
    else:
      arrsum[i] = float(temp[0])
      numstr.append(1)

    arrstr[i] = arrstr[i].strip()
    if arrstr[i][-2:-1].lower() in ('x', 'х') and arrstr[i][-1:].isdigit():
      if bisum:
        arrstr[i] = arrstr[i][:-2].strip()
      else:
        iswrong = True
    elif bisum:
        isWeight[i] = True

    
  return arrstr, arrsum, numstr, iswrong, isWeight
```

```python
wb2 = openpyxl.Workbook()
wb2.create_sheet(title = 'Первый лист', index = 0)
sheet2 = wb2['Первый лист']
sheet2.append(['Дата', 'Тип', 'Название', 'Цена', 'Кол-во/вес', 'Есть проблемы?', 'Это вес?'])

wbfin = openpyxl.Workbook()
wbfin.create_sheet(title = 'Покупки', index = 0)
sheetfin = wbfin['Покупки']
sheetfin.append(['Дата', 'Тип', 'Название', 'Цена', 'Кол-во/вес', 'Стоимость', 'Скидка', 'Магазин'])

num = 0
clms = list(sheet.columns)
for i in range(11, len(list(clms)) ):
  clm = list(clms[i])
  if not clm[0].value in (None, ' '):
    year = int(clm[0].value)
  if not clm[1].value in (None, ' '):
    month = monthDic[clm[1].value]
  day = int(clm[2].value)
  num += 1
  date = f'{year}.{month}.{day}'

  for j in range(0, 17, 2):
    arrstr, arrsum, numstr, iswrong, isWeight = str2arr(clm[3+j].value, clm[4+j].value)
    
    
    if not iswrong:
      if numstr:
        for w in isWeight:
          if w:
            break
      else:
        continue
    for k in range(len(numstr)):
      if iswrong or isWeight[k]:
        sheet2.append([date, types[j], arrstr[k], arrsum[k], numstr[k], f'{iswrong} в {i}.{j}', isWeight[k]])
      sheetfin.append([date, types[j], arrstr[k], arrsum[k], numstr[k], arrsum[k]*numstr[k], 0, 'Null'])

print(num)
wb2.save('forWeight.xlsx')
wbfin.save('All.xlsx')
```
